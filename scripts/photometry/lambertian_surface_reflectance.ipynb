## Lambertian Surface Reflectance
**Objective**: Simulate the reflectance of a Lambertian surface.

**Lambert law**:
$I = A \cos(\theta)$

( $\theta$ ) : Emergence

( $A$ ) : Albedo


```python
import numpy as np
import matplotlib.pyplot as plt

def lambertian_reflectance(A, theta_deg):
    theta_rad = np.deg2rad(theta_deg)
    return A * np.cos(theta_rad)

A = 0.5 
A_RGB = {
    'red': 0.35, # 650 nm (red)
    'green': 0.25, # 550 nm (green)
    'blue': 0.15, # 450 nm (blue)
}
angles = np.linspace(-90, 90, 100)

fig = plt.figure()

for ind, color in enumerate(list(A_RGB)):
    reflectance = lambertian_reflectance(A_RGB[color], angles)
    plt.plot(np.deg2rad(angles), reflectance, color=color)

plt.xlabel('Incidence Angle (degrees)')
plt.ylabel('Reflectance')
plt.title('Lambertian Reflectance Model')
plt.grid(True)
plt.xlim(np.deg2rad(-90),np.deg2rad(90))
plt.show()
```
